```python
import re
import pandas as pd


cars = pd.read_csv("car_price.xls")
cars["doornumber"] = cars["doornumber"].replace({"two": 2, "four": 4})
cars["car_company"] = cars["CarName"].apply(
    lambda x: re.findall(re.compile(r"\w+-?\w+"), x)[0]
)
# volkswagen
cars.loc[
    (cars["car_company"] == "vw") | (cars["car_company"] == "vokswagen"), "car_company"
] = "volkswagen"

# porsche
cars.loc[cars["car_company"] == "porcshce", "car_company"] = "porsche"

# toyota
cars.loc[cars["car_company"] == "toyouta", "car_company"] = "toyota"

# nissan
cars.loc[cars["car_company"] == "Nissan", "car_company"] = "nissan"

# mazda
cars.loc[cars["car_company"] == "maxda", "car_company"] = "mazda"
cars = cars.drop("enginetype", axis=1)
# cars = cars.select_dtypes(exclude=["object"])
cars.drop(["car_ID", "fuelsystem"], axis=1, inplace=True)
cars["symboling"] = cars["symboling"].astype(str)
cars = pd.get_dummies(
    cars,
    columns=cars.select_dtypes(include=["object"]).columns,
    drop_first=True,
    dtype=int,
)
cars["price_thousands_USD"] = cars["price"] / 1000
cars.drop(["price"], axis=1, inplace=True)
cars["price_thousands_USD"]
cars.to_csv("car_price.csv", index=False)
cars
```

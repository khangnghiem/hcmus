```python
import os
import pytesseract
from pdf2image import convert_from_path
from PIL import Image, ImageFilter
import shutil
import sys
import numpy as np
from scipy.ndimage import binary_opening
```

```python
# Convert PDF to images (PNG format)
pages = convert_from_path("../data/test_bia.pdf", 300)
png_files = []
for i, page in enumerate(pages):

    # Convert to grayscale
    page = page.convert("L")
    # Apply median filter to reduce noise
    # page = page.filter(ImageFilter.MedianFilter(size=3))
    for factor in [1, 1.2, 1.5, 1.8, 2, 2.5]:
        for threshold in range(128, 161, 2):
            page = page.point(
                lambda x: x // factor if x < threshold else 255 - ((255 - x) // factor)
            )
            png_filename = (
                f"../png/test_page_{i+1}_factor_{factor}_threshold_{threshold}.png"
            )
            page.save(png_filename, "PNG")
            png_files.append(png_filename)
```

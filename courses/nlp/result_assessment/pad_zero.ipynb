```python
import json
import os
import polars as pl
import shutil
import sys
```

```python
dir_name = "Han Nom"
dir_name = "PHIEN AM"
for filename in os.listdir(dir_name):
    name_part = filename.split('.')[0]
    prefix = name_part.split('-')[0]
    if len(prefix) < 5:
        prefix = prefix.zfill(5)
    new_name = prefix + '-' + name_part.split('-')[1] if '-' in name_part else prefix
    os.rename(os.path.join(dir_name, filename), os.path.join(dir_name, new_name + os.path.splitext(filename)[1]))
```

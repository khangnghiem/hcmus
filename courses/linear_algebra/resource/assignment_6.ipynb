# Exercise 6 - Phone Price Prediction Using Linear Regression and SVD

Linear regression model to predict phone prices based on features such as RAM, Memory, Battery, and a constant term. The dataset consists of 24 samples with features in matrix `A` and corresponding prices in vector `Y`.

## Workflow Overview

1. **Data Loading**: 
    - Matrix `A` contains the feature vectors for each phone (e.g., [RAM, Memory, Battery, 1]).
    - Vector `Y` holds the actual prices.

2. **Matrix Operations**:
    - Compute `AtA = A.T @ A` (the Gram matrix).
    - Calculate the determinant of `AtA` to check if it's invertible.

3. **Regression Coefficients**:
    - If `det(AtA) != 0`, use the normal equation: `X = (AtA)^(-1) @ A.T @ Y`.
    - If `det(AtA) ≈ 0` (singular matrix), apply SVD decomposition:
      - Decompose `A = U @ S @ Vt`.
      - Construct the pseudoinverse `Σ⁺` from `S`.
      - Compute `X = Vt.T @ Σ⁺ @ U.T @ Y`.

4. **Prediction**:
    - For a new phone with features `omega` (e.g., [4, 64, 4000, 1]), predict the price as `predicted_price = omega @ X`.

```python
# Step 1: Library
import numpy as np
from scipy.linalg import svd
```

```python
# Format: [RAM, Memory, Battery, 1]
A = np.array(
    [
        [1, 16, 3000, 1],
        [3, 32, 4000, 1],
        [4, 64, 5000, 1],
        [3, 32, 3300, 1],
        [3, 32, 5000, 1],
        [6, 128, 5000, 1],
        [4, 64, 4000, 1],
        [6, 128, 5000, 1],
        [8, 128, 4500, 1],
        [8, 256, 5000, 1],
        [8, 128, 4500, 1],
        [8, 128, 4100, 1],
        [8, 256, 4500, 1],
        [8, 128, 4500, 1],
        [8, 128, 4500, 1],
        [12, 128, 5000, 1],
        [12, 256, 4500, 1],
        [8, 256, 3300, 1],
        [8, 256, 3300, 1],
        [8, 256, 4800, 1],
        [12, 256, 4500, 1],
        [12, 256, 5000, 1],
        [12, 512, 5000, 1],
        [12, 512, 4400, 1],
    ],
    dtype=float,
)

# Prices vector Y
Y = np.array(
    [
        1_850_000,
        2_650_000,
        3_350_000,
        3_790_000,
        4_250_000,
        4_700_000,
        4_150_000,
        5_150_000,
        6_600_000,
        10_100_000,
        10_500_000,
        12_400_000,
        13_650_000,
        12_790_000,
        15_500_000,
        16_000_000,
        18_990_000,
        19_350_000,
        20_990_000,
        23_000_000,
        23_000_000,
        29_800_000,
        29_990_000,
        33_990_000,
    ],
    dtype=float,
)


# Step 2: Compute determinant (Định thức) of AtA
AtA = A.T @ A

det_AtA = np.linalg.det(AtA)

print(f"Determinant of A^T A: {det_AtA}\n")

# Step 2: SVD of A
if np.abs(det_AtA) > 1e-10:  # not 0
    X = np.linalg.inv(AtA) @ A.T @ Y
else:
    # Step 2.1 SVD
    U, S, Vt = svd(A)
    # Step 2.2 Build Σ⁺
    Sigma_plus = np.zeros((Vt.shape[0], U.shape[0]))
    for i in range(len(S)):
        if S[i] > 1e-10:
            Sigma_plus[i, i] = 1.0 / S[i]
    # Step 2.3 Compute Regression coefficients X = (pseudoinverse of A) @ Y = V @ Σ+ @ U.T @ Y
    X = Vt.T @ Sigma_plus @ U.T @ Y

print(f"Regression coefficients (X): {X}\n")

# Step 3: Predict new phone price
omega = np.array([4, 64, 4000, 1])
predicted_price = omega @ X

print(f"Predicted price for phone [4, 64, 4000, 1]: {predicted_price:,.0f}")
```

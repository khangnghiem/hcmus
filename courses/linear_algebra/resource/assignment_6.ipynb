```python
# Step 1: Library
import numpy as np
from scipy.linalg import svd
```

```python
# -----------------------------
# Step 1: Load dataset
# Format: [RAM, Memory, Battery, 1]
A = np.array(
    [
        [1, 16, 3000],
        [3, 32, 4000],
        [4, 64, 5000],
        [3, 32, 3300],
        [3, 32, 5000],
        [6, 128, 5000],
        [4, 64, 4000],
        [6, 128, 5000],
        [8, 128, 4500],
        [8, 256, 5000],
        [8, 128, 4500],
        [8, 128, 4100],
        [8, 256, 5000],
        [8, 128, 4500],
        [8, 128, 4100],
        [8, 256, 4500],
        [8, 128, 4500],
        [8, 128, 4500],
        [12, 128, 5000],
        [12, 256, 4500],
        [8, 256, 3300],
        [8, 256, 3300],
        [8, 256, 4800],
        [12, 256, 4500],
        [12, 256, 5000],
        [12, 512, 5000],
        [12, 512, 4400],
    ],
    dtype=float,
)

# Prices vector Y
Y = np.array(
    [
        1_850_000,
        2_650_000,
        3_350_000,
        3_790_000,
        4_250_000,
        4_700_000,
        4_150_000,
        5_150_000,
        6_600_000,
        10_100_000,
        10_500_000,
        12_400_000,
        13_650_000,
        12_790_000,
        15_500_000,
        16_000_000,
        18_990_000,
        19_350_000,
        20_990_000,
        23_000_000,
        23_000_000,
        29_800_000,
        29_990_000,
        33_990_000,
    ],
    dtype=float,
)

# Step 2: SVD of A
U, S, Vt = svd(A)

# Build Σ⁺
Sigma_plus = np.zeros((Vt.shape[0], U.shape[0]))
for i in range(len(S)):
    if S[i] > 1e-10:
        Sigma_plus[i, i] = 1.0 / S[i]

# Compute pseudoinverse of A
A_plus = Vt.T @ Sigma_plus @ U.T

# Regression coefficients
X = A_plus @ Y

print("Regression coefficients (X):")
print(X)

# -----------------------------
# Step 3: Predict new phone price
new_phone = np.array([4, 64, 4000, 1])
predicted_price = new_phone @ X

print("\nPredicted price for phone [4, 64, 4000]:")
print(predicted_price)
```
